```python
from langchain_groq import ChatGroq
from langchain.tools import tool
import os

# -----------------------
# 1ï¸âƒ£ Define Tool Properly
# -----------------------
@tool
def get_weather(city: str) -> str:
    """
    Get the current weather for a given city.

    Args:
        city: Name of the city (e.g., Delhi, Mumbai)

    Returns:
        A weather description string.
    """
    return f"The weather in {city} is sunny."


# -----------------------
# 2ï¸âƒ£ Initialize Groq LLM
# -----------------------
llm = ChatGroq(
    model="llama-3.1-8b-instant",   # correct lowercase
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY")
)


# -----------------------
# 3ï¸âƒ£ Bind Tool (FASTER than agent)
# -----------------------
llm_with_tools = llm.bind_tools([get_weather])


# -----------------------
# 4ï¸âƒ£ Invoke Model
# -----------------------
response = llm_with_tools.invoke(
    "What is the weather in Delhi?"
)

# -----------------------
# 5ï¸âƒ£ Print Clean Output
# -----------------------
print(response.content)
```

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain.tools import tool

load_dotenv()

@tool
def get_weather(city: str) -> str:
    """Get the current weather for a given city."""
    return f"The weather in {city} is sunny."

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY")
)

llm_with_tools = llm.bind_tools([get_weather])

# Step 1: Ask model
response = llm_with_tools.invoke(
    "What is the weather like in New York?"
)

# Step 2: Check if tool was called
if response.tool_calls:
    tool_call = response.tool_calls[0]
    tool_name = tool_call["name"]
    tool_args = tool_call["args"]

    # Step 3: Execute tool manually
    if tool_name == "get_weather":
        result = get_weather.invoke(tool_args)

        print("\nâœ… Final Answer:")
        print(result)

else:
    print("\nâœ… Direct Answer:")
    print(response.content)
```
